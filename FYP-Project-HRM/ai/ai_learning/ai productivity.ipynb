```python
import pandas as pd

courses_df = pd.DataFrame(courses)

courses_df.head()
```

```python
course_embeddings = model.encode(courses_df["skills"].tolist())
```

```python
def recommend_courses(user_skills, top_k=5):

    # convert skills list → sentence
    user_text = " ".join(user_skills)

    # generate user embedding
    user_embedding = model.encode([user_text])

    # compute similarity
    similarity_scores = cosine_similarity(user_embedding, course_embeddings)[0]

    # add scores to dataframe
    courses_df["score"] = similarity_scores

    # sort courses
    recommended = courses_df.sort_values("score", ascending=False)

    return recommended.head(top_k)
```
