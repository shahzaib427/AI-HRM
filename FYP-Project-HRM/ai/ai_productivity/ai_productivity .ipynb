```python
import pandas as pd
import numpy as np
```

```python
def calculate_productivity(data):

    productive = data["productiveHours"]
    meetings = data["meetings"]
    breaks = data["breaks"]
    distractions = data["distractions"]
    deep_work = data["deepWork"]

    score = (
        productive * 10 +
        deep_work * 15 -
        distractions * 20 -
        meetings * 5 -
        breaks * 3
    )

    score = max(0, min(100, score))

    return round(score,2)
```

```python
def generate_ai_suggestions(data):

    suggestions = []

    if data["distractions"] > 0.5:
        suggestions.append({
            "title":"Disable notifications",
            "impact":"Reduce distractions by 60%"
        })

    if data["meetings"] > 1.5:
        suggestions.append({
            "title":"Reduce meeting duration",
            "impact":"Save 3 hours weekly"
        })

    if data["deepWork"] < 4:
        suggestions.append({
            "title":"Schedule deep work blocks",
            "impact":"Increase focus by 25%"
        })

    return suggestions
```

```python
data = {
    "productiveHours": 6.5,
    "meetings": 2,
    "breaks": 1.5,
    "distractions": 0.8,
    "deepWork": 3.2
}

score = calculate_productivity(data)
suggestions = generate_ai_suggestions(data)

print("Score:",score)
print("Suggestions:",suggestions)
```
